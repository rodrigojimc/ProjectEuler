```python
class message () :
    
    def __init__(self, original): 
        
        # Has been decoded at least once
        self.decoded = False
        
        # Key has been given at leats once
        self.key_given = False
        
        # Last key, None by default
        self.key_text = None
        self.key_dec = None
        self_key_bin = None
        
        # Last decodification, None by default
        self.decoded_text = None
        self.decoded_dec = None
        self_decoded_bin = None
        
        
        # original text input conversions
        if type(original) == str:
            
            self.original_text = original
            self.original_dec = self.text_to_dec(original)
            
        elif type(original) == list or type(original) == tuple:
            
            valid = True
            
            for x in original:
                
                if type(x) != int : valid = False 
                elif x > 256 : valid = False 
                    
            if valid:
                
                self.original_dec = original
                self.original_text = self.dec_to_text(original)
                
            else: print("Error: list/tuple must contain just integers between 1 and 255")
            
        else: print("Error, input should be string or list/tuple of integers ")
            
        self.original_bin = self.dec_to_bin(self.original_dec)
     
    #########################
    #  Coversion Functions  #
    #########################
    
    def dec_to_text (self, d):
        return ("".join(chr(x) for x in d))
    
    def text_to_dec (self, t):
        return ([ord(x) for x in t])
    
    def dec_to_bin (self,d):
        return (["0"*(8-math.floor(math.log(x+0.5,2)))+"".join(bin(x)[2:]) for x in d])
    
    def bin_to_dec (self, b):
        v=[]
        for x in b:
            d=0
            l=len(x)
            for i in range(l):
                d+=int(x[i])*2**(l-i-1)
            v.append(d)
        return v
    
    #######################
    #   Print Functions   #
    #######################
    
    def print_original (self):
        print(self.original_text)
        print(self.original_dec)
        print(self.original_bin)
        
    def print_decoded (self):
        if self.decoded == True:
            print(self.decoded_text)
            print(self.decoded_dec)
            print(self.decoded_bin)
        else: print("Hasn't been decoded yet")
            
    def print_key (self):
        if self.key_given == True:
            print(self.key_text)
            print(self.key_dec)
            print(self.key_bin)
        else: print("Key hasn't been given")

    ######################
    #  Decode functions  #
    ######################

    def decode (self, key = None):
        
        if key != None : 
            self.key_gen(key)
            
        if self.key_given == True:
            
            if key == None : key = self.key_dec
        
            self.decoded = True
            self.key_gen(key)
        
            lm=len(self.original_bin)
            lk=len(self.key_bin)
            self.decoded_bin = [self.xor(self.original_bin[i], self.key_bin[i%lk]) for i in range (lm)]
            
            self.decoded_dec = self.bin_to_dec(self.decoded_bin)
            self.decoded_text = self.dec_to_text(self.decoded_dec)
            
            #self.print_decoded()
            
            return self.decoded_text
            
        else: print("Key hasn't been given")
        
    def xor (self, a, b):
        c=""   
        for i in range (9):
            c+="".join(str(int(a[i])^int(b[i])))   
        return (c)
    
    # Key input conversions
    def key_gen (self, key):
        self.key_given = True
            
        if type(key) == str:
            
            self.key_text = key
            self.key_dec = self.text_to_dec(key)
            
        elif type(key) == list or type(key) == tuple:
            
            valid = True
            
            for x in key:
                
                if type(x) != int : valid = False 
                elif x > 256 : valid = False 
                    
            if valid:
                
                self.key_dec = key
                self.key_text = self.dec_to_text(key)
                
            else: print("Error: list/tuple must contain just integers between 1 and 255")
            
        else: print("Error, input should be string or list/tuple of integers ")
            
        self.key_bin = self.dec_to_bin(self.key_dec)
    
        
```
