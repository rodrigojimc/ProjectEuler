```python
import math as m
```

```python
def sumdp (n) :
    v=[1]*(n+1)
    f=m.floor(n/2)+1
    for d in range (2,f) :
        l=m.floor(n/d)
        for i in range (2,l+1) :
            v[i*d]=v[i*d]+d
    return v
```

```python
v=sumdp(28123)
```

```python
v[0]=0
for i in range(len(v)):
    if v[i]<=i : v[i]=0
```

```python
ab=[]
for i in range(len(v)):
    if v[i]!=0 : ab.append(i)
```
