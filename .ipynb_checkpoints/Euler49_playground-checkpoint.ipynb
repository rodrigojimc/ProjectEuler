```python
def perm_d (n): return set(sorted([int("".join(str(d) for d in p)) for p in perm([int(d) for d in str(n)])]))
```

```python
def recursion (m,n):
    v=[]
    for i in range(0,m):
        v.append(i+1)
    print(v)
    if m<n:
        recursion (m+1,n)  
```
