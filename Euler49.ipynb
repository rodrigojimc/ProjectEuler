```python
def dif(v,n=0,M=[]):
    if n==0: M=[]
    l=len(v)
    D=[]
    for i in range (n+1,l):
        d=v[i]-v[n]
        D.append(d)
    M.append(D)
    if n+2<l:
        M=dif(v,n+1,M)
    return (M)

def find_as (v):
    y=(0,0,0)
    l=len(v)
    for i in range (l):
        for j in range (l-i-1):
            for k in range (len(v[i+j+1])):
                if v[i][j]==v[i+j+1][k] : 
                    y=(i,i+j+1,i+j+k+2)
    return y
```

```python
import time
from sympy import sieve
from itertools import permutations as perm

St=time.time()

def perm_d (n): return tuple(set(sorted([int("".join(str(d) for d in p)) for p in perm([int(d) for d in str(n)])])))

def dif(v,n=0,M=[]):
    if n==0: M=[]
    l=len(v)
    D=[]
    for i in range (n+1,l):
        d=v[i]-v[n]
        D.append(d)
    M.append(D)
    if n+2<l:
        M=dif(v,n+1,M)
    return (M)

def find_as (v):
    y=(0,0,0)
    l=len(v)
    for i in range (l):
        for j in range (l-i-1):
            for k in range (len(v[i+j+1])):
                if v[i][j]==v[i+j+1][k] : 
                    y=(i,i+j+1,i+j+k+2)
    return y

primes=list(sieve.primerange(1000,10000))

v=[x for x in sorted(set([tuple(sorted([p for p in perm_d(n) if p in primes])) for n in primes])) if len(x)>2]

l=len(v)
for i in range (l):
    y=find_as(dif(v[i]))
    if y!=(0,0,0):
        print("(a,b,c)=",(v[i][y[0]],v[i][y[1]],v[i][y[2]]),"| diference=",v[i][y[1]]-v[i][y[0]])
        
Et=time.time()-St    
print("elapsed time:",Et)
```
